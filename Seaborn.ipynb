```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('My Cricket Dataset.xlsx')
```

```python
sns.histplot(df['runs'], kde=True)
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=df, x='balls_faced', y='runs', hue='player_name')
plt.title('Runs vs Balls Faced')
plt.xlabel('Balls Faced')
plt.ylabel('Runs')
plt.show()
```

```python
sns.barplot(data=df, x='player_name', y='runs')
plt.title('Total Runs by Player')
plt.xlabel('Player Name')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=df, x='player_name', y='wickets')
plt.title('Total Wickets by Player')
plt.xlabel('Player Name')
plt.ylabel('Wickets')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=df, x='player_name', y='balls_bowled')
plt.title('Total Balls Bowled by Player')
plt.xlabel('Player Name')
plt.ylabel('Balls Bowled')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df, vars=['runs', 'balls_faced', 'wickets', 'balls_bowled'], hue='player_name')
plt.show()
```

```python
sns.jointplot(data=df, x='balls_faced', y='runs', kind='reg')
plt.show()
```

```python
sns.violinplot(data=df, x='player_name', y='runs')
plt.title('Distribution of Runs by Player')
plt.xlabel('Player Name')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_matrix = df[['runs', 'balls_faced', 'wickets', 'balls_bowled']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.set(style='whitegrid')  # Setting the style
plt.figure(figsize=(10, 6))  # Setting the figure size

sns.scatterplot(data=df, x='balls_faced', y='runs', hue='player_name')
plt.title('Runs vs Balls Faced')
plt.xlabel('Balls Faced')
plt.ylabel('Runs')
plt.show()
```
